```python
path = '/rasmussen-scratch/mrocque/research/relampago/chivo_1bf_more/rhi/20181214/'
file = 'chivo.1bf.20181214_030708.REL_RHI45.nc'

radar = pyart.io.read_cfradial(path+file)
display = pyart.graph.RadarDisplay(radar)
```

```python
print (radar.fields.keys())
```

```python
plt.figure(figsize=(10,5))

display.plot('DZ_qc', 0, vmin=0, vmax=70, title_flag=False, cmap='pyart_HomeyerRainbow')
display.set_limits([0, 150], [0,20])
```
